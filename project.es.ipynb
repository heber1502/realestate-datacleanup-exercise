#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
ds = pd.read_csv('assets/real_estate.csv', sep=';')
df = pd.DataFrame(ds)

# Lista de poblaciones del cinturón sur
cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame para las poblaciones del cinturón sur
df_sur = df[df['level5'].isin(cinturon_sur)]

# Crear un diccionario para almacenar los datos normalizados por población
datos_normalizados = {}

# Normalizar la variable de precios por cada población
for poblacion in cinturon_sur:
    # Filtrar por población
    df_poblacion = df_sur[df_sur['level5'] == poblacion]
    
    # Calcular media y desviación estándar
    media = df_poblacion['price'].mean()
    desviacion_estandar = df_poblacion['price'].std()
    
    # Normalizar la variable de precios
    precios_normalizados = (df_poblacion['price'] - media) / desviacion_estandar
    
    # Almacenar los precios normalizados en el diccionario
    datos_normalizados[poblacion] = precios_normalizados

# Crear un gráfico con histogramas para cada población
plt.figure(figsize=(12, 8))

for poblacion, precios_normalizados in datos_normalizados.items():
    plt.hist(precios_normalizados, bins=30, alpha=0.5, label=poblacion)

plt.title('Histogramas de Precios Normalizados por Población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(loc='upper right')
plt.show()
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
import pandas as pd

# Leer el archivo CSV
ds = pd.read_csv('assets/real_estate.csv', sep=';')
df = pd.DataFrame(ds)

# Lista de municipios de interés
municipios_interes = ["Getafe", "Alcorcón"]

# Filtrar el DataFrame para los municipios de interés
df_interes = df[df['level5'].isin(municipios_interes)].copy()

# Crear la nueva columna pps (precio por metro cuadrado) usando .loc[]
df_interes.loc[:, 'pps'] = df_interes['price'] / df_interes['surface']

# Filtrar datos por municipio usando el método query
df_getafe = df_interes.query("level5 == 'Getafe'")
df_alcorcon = df_interes.query("level5 == 'Alcorcón'")

# Calcular estadísticas descriptivas para pps en ambos municipios
estadisticas_getafe = df_getafe['pps'].describe()
estadisticas_alcorcon = df_alcorcon['pps'].describe()

# Imprimir las estadísticas
print("Estadísticas de Precio por Metro Cuadrado en Getafe:")
print(estadisticas_getafe)
print("\nEstadísticas de Precio por Metro Cuadrado en Alcorcón:")
print(estadisticas_alcorcon)

# Comparar los precios por metro cuadrado
precio_promedio_getafe = estadisticas_getafe['mean']
precio_promedio_alcorcon = estadisticas_alcorcon['mean']

print(f"\nPrecio promedio por metro cuadrado en Getafe: {precio_promedio_getafe:.2f}")
print(f"Precio promedio por metro cuadrado en Alcorcón: {precio_promedio_alcorcon:.2f}")

if precio_promedio_getafe > precio_promedio_alcorcon:
    print("El precio por metro cuadrado en Getafe es mayor que en Alcorcón.")
else:
    print("El precio por metro cuadrado en Alcorcón es mayor o igual que en Getafe.")
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
ds = pd.read_csv('assets/real_estate.csv', sep=';')
df = pd.DataFrame(ds)

# Lista de poblaciones de interés
cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame para las poblaciones de interés
df_sur = df[df['level5'].isin(cinturon_sur)].copy()

# Crear la nueva columna pps (precio por metro cuadrado) en df_sur
df_sur['pps'] = df_sur['price'] / df_sur['surface']

# Crear un gráfico con subplots para cada población
fig, axs = plt.subplots(2, 2, figsize=(14, 10), sharex=True, sharey=True)
fig.suptitle('Diagrama de Dispersión de Precio por Metro Cuadrado por Población')

# Iterar sobre las poblaciones e índices de subplots
for i, poblacion in enumerate(cinturon_sur):
    ax = axs[i // 2, i % 2]  # Determinar la posición del subplot
    df_poblacion = df_sur[df_sur['level5'] == poblacion]
    
    ax.scatter(df_poblacion['surface'], df_poblacion['pps'], alpha=0.7)
    ax.set_title(poblacion)
    ax.set_xlabel('Superficie (m²)')
    ax.set_ylabel('Precio por m²')

# Ajustar el diseño para que no se sobrepongan los elementos
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
from ipyleaflet import Map, basemaps, Marker, MarkerCluster, DivIcon
import pandas as pd

# Leer el archivo CSV
ds = pd.read_csv('assets/real_estate.csv', sep=';')
df = pd.DataFrame(ds)

# Lista de poblaciones de interés
cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame para las poblaciones de interés
df_sur = df[df['level5'].isin(cinturon_sur)].copy()

# Crear la nueva columna pps (precio por metro cuadrado) en df_sur
df_sur['pps'] = df_sur['price'] / df_sur['surface']

# Crear un diccionario para almacenar coordenadas
coordenadas_dict = {}

# Iterar sobre cada población e ingresar coordenadas al diccionario
for poblacion in cinturon_sur:
    df_poblacion = df_sur[df_sur['level5'] == poblacion]
    coordenadas = df_poblacion[['latitude', 'longitude']].values.tolist()
    coordenadas_dict[poblacion] = coordenadas

# Mapa centrado en Madrid con OpenStreetMap
map = Map(center=(40.4168, -3.7038), zoom=10, min_zoom=1, max_zoom=20, 
          basemap=basemaps.OpenStreetMap.Mapnik)

# Colores para cada población
colores = {
    "Fuenlabrada": "red",
    "Leganés": "blue",
    "Getafe": "green",
    "Alcorcón": "purple"
}

# Agregar marcadores al mapa
for poblacion, coordenadas in coordenadas_dict.items():
    markers = [
        Marker(
            location=(lat, lon),
            icon=DivIcon(
                html=f'<div style="background-color:{colores[poblacion]}; width: 10px; height: 10px; border-radius: 50%;"></div>'
            )
        ) for lat, lon in coordenadas
    ]
    marker_cluster = MarkerCluster(markers=markers)
    map.add_layer(marker_cluster)

map
```
